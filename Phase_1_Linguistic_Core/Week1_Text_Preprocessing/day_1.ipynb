## **Part 1 - What is Text Normalization ?**

### **The Core Idea**
Text Normalization = Converting raw and meesy text into a standard format

Real world text is messy.

#### Example raw data: 
- OMG!!! I LOOOOVE AI 🤖🔥
- Visit https://ai.com NOW!!!
- Price is $1000!!!!

#### Machine see this as:
- OMG
- !!!
- LOOOOVE
- 🤖
- 🔥
- https://ai.com
- $1000

But what we actually want is somthing like:
**love ai visit url price number**

---

## Why Text Normalization Exists

Because lanugage variability is huge.

#### Example:
- love
- Love
- LOVE
- loooove

Humans understand these as similar.

But NLP model sees **four completely different tokens**.

Normalization **reduces noise**.

---

## Vocabulary Explosion Problem

Imagine a dataset:

- Apple 
- apple
- APPLE
- apple!!!

**Without normalization: Vocabulary size = 4**

**With normalization: Vocabulary size = 1**

This improves:
- model learning
- statistical reliability
- memory efficieny

---

## When NOT to Normalize Too Much

Modern transfomer model like:
- Bert
- GPT

Often prefers **minimal processing**

#### Example:
- Apple -> company
- apple -> fruit

Lowercasing removes this distinction.

This is why there are models like:

- bert-base-cased
- bert-base-uncased

---

## Core Normalization Steps

Typical Classical NLP Pipeline:

- Raw text 
- Lowercase 
- Remove punctuation(marks used for dividing writing into sentences and phrases eg. periods, commas, and question marks) 
- Remove stopwords(words contribute little to meaning of sentence eg. a, an, the) 
- Stemming(reducing words to their stem, may not always be a valid word) / Lemmatization(reducing words to their dictionary form, consdering context for accurate interpretation) 
- Feature extraction

## **Part - 2 Lowercasing**

### **Theory**

Lowercasing converts all text to lowercase.

#### Example: "I Love AI" to "i love ai"

---

## Why Lowercasing works

Language has **case variations**.

#### Example:
- Apple
- apple
- APPLE

Without lowercasing: Vocbulary = 3

With lowercasing: Vocabulary = 1

This improves:
- statistical learning
- frequency estimation
- token consistency

---

## Practical 1 - Lowercasing in Python

```python
text = "I Love NLP and AI"

normalized = text.lower()

print("Output : ",text)
```

## Practical 2 - Lowercasing a Dataset

```python
sentences = [
    "I LOVE NLP",
    "NLP is AMAZING",
    "I enjoy Learning AI"
]

lower_sentences = [s.lower() for s in sentences]

print("Output : ", lower_sentences)
```

## Practial 3 - Using spaCy

#### **Using spaCy tokens.**

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Apple is Launching New AI Products"

doc = nlp(text)

print("Output :")
for token in doc:
    print(token.text, "->", token.lower_)
```

## Practical 4 - Using Regex Pipeline

#### We'll combine **lowercasing + cleaning**

```python
import re

text = "OMG!!! I LOVE AI!!!"

text = text.lower()

text = re.sub(r"[^\w\s]", "", text) 

print("Output :", text, end="\n\n")
print("[^\w\s] means remove anything that is not:","\n","- word character", "\n", "- space")
```

## Mini Experiment

#### This shows vocabulary shrink and **why preprocessing matters**

```python
text = [
"Apple",
"apple",
"APPLE",
"Apple!",
"APPLE!!!"
]

print("Without normalization:", set(text))

normalized = [t.lower() for t in text]

print("With normalization:", set(normalized))
```
