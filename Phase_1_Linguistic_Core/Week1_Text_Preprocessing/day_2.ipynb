---

## **Part 1 - Why Cleaning Text Matters**

#### **Real World Problem**

Raw text is noisy : "Hello!!! @john, how are you??? #AI is awesome!!! 💯🔥"

#### For a machine this becomes:

- Hello
- !!!
- john
- ???
- #AI
- 💯
- 🔥

#### This symbols:
- don't add meaningful semantic value(most of the time)
- increase vocabulary size
- confuse statistical model

---

## **Goal of Today**

**Convert**: "Hello!!! @john #AI is awesome 💯🔥"   **to** : hello john ai is awesome

---

## **Important Insight**

#### Not all punctuation is useless:

#### Example:
- "Let's eat, grandma"
- "Let's eat gradma"

Meaning changes completely, So cleaning is **task-dependent** 

---

## **Part 2 - Types of Noise**

#### We mainly deal with:

**1 Punctuation :** !?.,:;'"

**2 Special Characters :** @#$%^&*()

**3 Social Media Noise :** @username #hastag

**4 Formatting Noise :** !!! ??? .....

---

## **Part 3 - Regex**

#### Regex = patter-based text manipulation, used everywhere in NLP.

#### **Most Important Pattern: [^\w\s]**

#### Meaning:
Removing everything that is NOT:
- \w -> letters + numbers
- \s -> spaces

---

## **Practial 1 - Basic Punctuation Removal**

```python
import re

text  = "Hello!!! How are you???"

clean = re.sub(r"[^\w\s]", "", text)

print("Output:", clean)
```

---

## **Practical 2 - With Lowercasing** 

```python
import re

text  = "Hello!! How are you???"

clean = re.sub(r"[^\w\s]","",text) # remove punctuation

print("Output:",clean)
```

---

## **Practical 3 - Handling @ and**

**Example:** @john loves #AI

**We want:** john loves ai


```python
import re 

text = '@john loves #AI'

text = text.lower()

text = re.sub(r"[@#]", "", text)  # remove mentions/hashtags

print("Output:", text)
```
