## **Part 1 - Why Cleaning Text Matters**

#### **Real World Problem**

Raw text is noisy : "Hello!!! @john, how are you??? #AI is awesome!!!"

#### For a machine this becomes:

- Hello
- !!!
- john
- ???
- #AI

#### This symbols:
- don't add meaningful semantic value(most of the time)
- increase vocabulary size
- confuse statistical model

---

## **Goal of Today**

**Convert**: "Hello!!! @john #AI is awesome"   **to** : hello john ai is awesome

---

## **Important Insight**

#### Not all punctuation is useless:

#### Example:
- "Let's eat, grandma"
- "Let's eat gradma"

Meaning changes completely, So cleaning is **task-dependent** 

---

## **Part 2 - Types of Noise**

#### We mainly deal with:

**1 Punctuation :** !?.,:;'"

**2 Special Characters :** @#$%^&*()

**3 Social Media Noise :** @username #hastag

**4 Formatting Noise :** !!! ??? .....

---

## **Part 3 - Regex**

#### Regex = patter-based text manipulation, used everywhere in NLP.

#### **Most Important Pattern: [^\w\s]**

#### Meaning:
Removing everything that is NOT:
- \w -> letters + numbers
- \s -> spaces

---

## **Practial 1 - Basic Punctuation Removal**

```python
import re

text  = "Hello!!! How are you???"

clean = re.sub(r"[^\w\s]", "", text)

print("Output:", clean)
```

---

## **Practical 2 - With Lowercasing** 

```python
import re

text  = "Hello!! How are you???"

clean = re.sub(r"[^\w\s]","",text) # remove punctuation

print("Output:",clean)
```

---

## **Practical 3 - Handling @ and**

**Example:** @john loves #AI

**We want:** john loves ai


```python
import re 

text = '@john loves #AI'

text = text.lower()

text = re.sub(r"[@#]", "", text)  # remove mentions/hashtags

print("Output:", text)
```

---

## **Practical 4 -  Removing Extra Symbols**

```python
import re 

text = "WOW!!!!!! This is amazing!!!!!!"

clean = re.sub(r"[!]+","!",text) # + remove one or more occurences, normalize repeated punctuation instead of removing it.

print("Output:",clean)
```

---

## **Practical 5 - Remove Everything Except Words**

```python
import re

text = "AI @2026 is #1!!!"

clean = re.sub(r"[^a-zA-Z\s]", "", text)

print("Output:",clean)
```

## **Part 4 - Using NLTK**

#### NLTK uses **tokenization + filtering.**

```python
import nltk
from nltk.tokenize import word_tokenize
import string

nltk.download("punkt") # downloads the “punkt” tokenizer data for the NLTK, a pre-trained sentence tokenizer, split text into sentences & words
nltk.download("punkt_tab")

text = "Hello!!! How are you???"

tokens = word_tokenize(text)

print("tokens:", tokens)

clean_tokens = [word for word in tokens if word not in string.punctuation]

print("Output:", clean_tokens)
```

---

## **Part 5 - Using spaCy**

#### **spaCy identifies punctuation tokens automatically.**

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text =  "Hello!!! How are you???"

doc = nlp(text)

print("doc:",doc)

tokens = [token.text for token in doc if not token.is_punct]

print("Ouput:", tokens)
```

## **Part 6 - Advanced Cleaning**

#### **Real pipelines combine multiple steps.**

```python
import re

def clean_text(text):
    text =  text.lower()

    #remove urls, start at http, grab everything until a space appears
    text = re.sub(r"http\S+", "", text)

    #remove @ and #
    text = re.sub(r"[@#]", "", text)

    #remove punctuation
    text = re.sub(r"[^\w\s]", "", text)

    #remove extraspaces
    text = re.sub(r"\s+", " ",text).strip()

    return text

text = "OMG!!! @john loves #AI Visit https://ai.com NOW!!!"

print("Output:", clean_text(text))
```

## **When NOT to Remove Punctuation**

#### - **1. Sentiment Analysis:** "I love this!!!" ("!!!" adds strong emotion)
#### - **2. Chatbots / LLMs:** Models like GPT use punctuation as context signals
#### - **3. Named Entities:** U.S.A, Removing punctuation is acceptable sometimes, but risky

---

## **Key Intuition**

#### Text cleaning removes:

#### **1. Noise:** !!! ??? ###
#### **2. Redundancy:** Hello!!! -> Hello
#### **3. Irrelevant symbols:** @ # $
