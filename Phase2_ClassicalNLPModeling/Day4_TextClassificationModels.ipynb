# 1. INTUITION

Imagine you're sorting emails into folders:
- If the email contains words like "win", "offer", "free" -> likely **spam**
- If it contains "meeting", "project", "deadline" -> likely **work**

You're not deeply understanding the sentence, you're just using **word patterns + probabilities**

That's exactly how **Naive Bayes text classification** works.

#### **What problem does this solve?**

- Automatically assign text to categories:
    - Spam vs not spam
    - Positive vs negative sentiment
    - Topic classification

#### **Why was this needed?**

- Manual classification doesn't scale
- Needed a **fast, simple, probabilistic model**
- Works well even with limited data

Key Idea:
**Use word probabilities to predict class labels**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Naive Bayes (Definition)**

A probabilistic classifier based on:
- Bayes' Theorem
- Assumption: **features (words) are independent** Each word is independent of other words (given the class)
- Feature independence changes, complex joint probability -> product of individual probabilities

---

**Intuition**

- Calculate probability of each class given words
- Choose class with highest probability

---

#### **2.2 How It Works**

1. Convert text -> vectors (TF-IDF/BoW)
2. Learn:
    - Probability of each class (or prior probab of each class, how common is this class, P(class))
    - Probability of each word given class (likelihood of words given class, P(word | class), eg. if this is spam, how likely is the word 'money'?)
3. For new text:
    - Compute probability of each class
    - Select highest

---

#### **2.3 Multinomial Naive Bayes**

**Definition**
- Uses **word counts or frequencies**

---

**Use Case**
- TF-IDF or BoW features
- Text classification tasks

---

**Example**

"win win offer" -> counts matter

---

#### **2.4 Bernoulli Naive Bayes**

**Definition**
- Uses **binary features (presence/absence)**

---

**Use Case**
- Whether a word exists or not

--- 

**Example**

"win win offer" -> treated same as "win offer"

---

#### **2.5 Mutinomial vs Bernoulli**

| **Feature** | **Multinomial** | **Bernoulli** |
|-------------|-----------------|---------------|
| Input | Counts/TF-IDF | Binary(0/1) |
| Focus | Frequency | Presence |
| Best for | Longer text | short text |

---

#### **2.6 Pipeline: TF-IDF -> Classifier**

**Flow**

1. Raw text 
2. Preprocessing
3. TF-IDF vectorization
4. Naive Bayes classifier
5. Prediction

---

**Example**

Training gave:

- P(spam) = 0.5
- P(win|spam) = 0.4
- P(money|spam) = 0.2

- P(ham) = 0.5
- P(win|ham) = 0.01
- P(money|ham) = 0.01

Predict: "win money"

Spam:
0.5 × 0.4 × 0.2 = 0.04

Ham:
0.5 × 0.01 × 0.01 = 0.00005

**Spam wins**

---

**Why this pipeline works**

- TF-IDF extracts important features
- Naive Bayes uses probabilities on those features

# 3. INTERVIEW NOTES
# =========================================

#### **Key Points**

- Naive Bayes is fast and effective for text classification
- Assumes features independence (not true, but works well)
- Multinomial is most commonly used for NLP

---

#### **Advantages**

- Simple and fast 
- Works well with high-dimensional data
- Performs well on small datasets

---

#### **Disadvantages**

- Independence assumption is unrealistic
- Cannot capture context or word relationships
- Limited performance compared to deep learning

---

#### **When to Use**

- Spam detection
- Sentiment analysis
- Topic classification
- Baseline models

---

#### **When NOT to Use**

- Complex semantic tasks
- Context heavy problems
- Large-scale deep NLP tasks

---

#### **Important Comparisons**

**Multinomial vs Bernoulli**

- Multinomial -> frequency based -> better for most NLP
- Bernoullo -> presence based -> useful for short text

| **Feature** | **Naive Bayes** | **Logistic Regression** |

# 4. IMPLEMENTATION
# =========================================

#### **4.1 Setup**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

---

#### **4.2 Sample Data**

```python
texts = [
    "Win money now",
    "Limited time offer",
    "Meeting at 5pm",
    "Project deadline tomorrow"
]

labels = ["spam", "spam", "ham", "ham"]
```

---

####  **4.3 Train-Test split**

```python
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.25, random_state=42
)
```

---

#### **4.4 Pipeline: TF-IDF + MultinomialNB**

```python
pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", MultinomialNB())
])

pipeline.fit(X_train, y_train)
y_preds = pipeline.predict(X_test)

print(classification_report(y_test,y_preds))
```

---

#### **4.5 BernoulliNB Variation**


```python
pipeline = Pipeline([
    ("tfidf",TfidfVectorizer(binary=True)),
    ("clf",BernoulliNB())
])

pipeline.fit(X_train,y_train)
```

---

#### **4.6 Prediction**

```python
print(pipeline.predict(["Free money offer"]))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
train_texts = [
    "Win a free iPhone now",
    "Limited time offer claim your prize",
    "Get cash rewards instantly",
    "Congratulations you have won a lottery",
    "Exclusive deal just for you click now",
    
    "Let's schedule a meeting tomorrow",
    "Project deadline is next week",
    "Can you review the document",
    "Lunch plan at 1pm",
    "Please call me when you are free"
]

train_labels = [
    "spam",
    "spam",
    "spam",
    "spam",
    "spam",
    
    "ham",
    "ham",
    "ham",
    "ham",
    "ham"
]

test_texts = [
    "free cash offer",
    "meeting tomorrow",
    "win money now",
    "project update needed",
    "Win a free iPhone",
    "Let's schedule a meeting",
    "Exclusive offer just for you"
]

test_labels = [
    "spam",
    "ham",
    "spam",
    "ham",
    "spam",
    "ham",
    "spam"
]

multinb_pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", MultinomialNB())
])

bernb_pipeline = Pipeline([
    ("tfidf", TfidfVectorizer(binary=True)),
    ("clf", BernoulliNB())
])
```

#### **MultinomialNB**

```python
multinb_pipeline.fit(train_texts, train_labels)
```

```python
multinb_pred = multinb_pipeline.predict(test_texts)

print(classification_report(test_labels, multinb_pred))
```

```python
for text in test_texts:
    print(text, ": ", multinb_pipeline.predict([text]))
```

---

#### **BernoulliNB**

```python
bernb_pipeline.fit(train_texts, train_labels)
```

```python
bernb_pred = bernb_pipeline.predict(test_texts)
print(classification_report(test_labels, bernb_pred))
```

```python
for text in test_texts:
    print(text,": ", bernb_pipeline.predict([text]))
```

# 8. SELF-TEST QUESTIONS
# =========================================

1. **Why does Naive Bayes work well despite the independence assumption?**

Naive Bayes works well despite the independence assumption because, although assumption is unrealistic, it simplifies the computation of probabilities in high-dimensional text data. In real language, words are dependent on each other, but modeling full join probabilities would enormous data and computation. By assuming independence, naive bayes estimates probabilities using individual word frequencies, which are easy to compute and robust even with limited data. In practice, many words still provide strong independent signals for classification (e.g. "offer", "free" for spam), so the model performs surprisingly well. This trade-off between simplicity and effectiveness makes Naive Bayes a strong baseline for text classification.

---

2. **When would you prefer BernoulliNB over MultinomialNB?**

BernoulliNB is preferred when the presence or absence of words is more important than their frequency, particularly in short text scenarios. In cases such as SMS classification, tweets or keyword detection, repeating a word multiple times doesn't add much additional information, so a binary representation is sufficient. BernoulliNB explicitly models whether a word appears or not, which makes it suitable for such tasks. In contrast, multinomialnb considers word counts or frequencies and generally performs better for longer documents where repetition carries meaningful information.

---

3. **Why TF-IDF commonly used for before Naive Bayes?**

TF-IDF is commonly used before naive bayes because it transforms raw text into a weighted numerical representation that emphasizes informative words while reducing the influence of common, less meaningful terms. Since naive bayes relies on feature probabilities, providing it with better-weighted features improves its ability to distinguish between classes. TF-IDF aligns well with naive bayes because both operates on individual features (words) withou modeling dependencies between them. As a result, the combination of TF-IDF and naive bayes creates and efficient and effective pipeline for keyword based classification tasks.
