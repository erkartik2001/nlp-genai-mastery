# =========================================
# 1. INTUITION
# =========================================

Imagine you're filtering emails into **spam** vs **not spam** using a checklist of signals:

- "free", "offer", "win" -> push toward spam
- "meeting", "project" -> push toward not spam

Each word contribute a **vote with some weight**, and you draw a **line** that separates spam from not spam.

- **Logistic Regression**: learns that line by combining weighted signals and outputs a **probability**
- **SVM**: draws a line that **maximizes the gap (margin)** between classes to make the separation more robust.

#### **What problem does this solve?**

- Classify text into categories using vector features (TF-IDF/embeddings)
- Learn a **decision boundary** that separates classes

#### **Why was this needed?**

- Naive Bayes ignore feature interactions and can be biased by independence assumption
- Needed models that:
    - Learn **optimal separating boundaries**
    - Work well in **high dimensional sparse spaces (text)**

Key Idea: 
**Turn text into vectors -> learn a boundary that separates classes**

---

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Logistic Regression (for text)**

**Definition**

A linear model that predicts the **probability** of a class using a weight sum of features followed by a sigmoid function.

---

**Intuition**

- Each word has weight
- Sum of (word * weight) -> score
- Convert score -> probability
- Threshold (e.g 0.5) -> class

---

**Decision Boundary**

- A **linear boundary (hyperplane)** in feature space
- Separate classes based on weighted combination of features

---

**Step-by-Step**

1. Text -> TF-IDF / Embeddings
2. Compute linear score: wx+b
3. Apply sigmoid -> probability
4. Classify based on threshold

---

#### **2.2 Decision Boundary Intuition**

- In 2D a line
- In higher dimensions a hyperplane
- Points on one side -> class A
- Other side -> class B

Better weights -> better separation

---

#### **2.3 Support Vector Machine (SVM)**

**Definition**

A model that finds the **maximum margin hyperplane** that best separates classes.

---

**Intuition**

- Don't just separate classes
- Separate them with the **widest possible gap**

---

**Key Concepts**

- **Support vectors**: closest points to the boundary
- **Margin**: distance between boundary and closest points
- **Objective**: maximize margin

---

#### **2.4 Linear vs Kernel SVM**

- **Linear SVM**: works well for text (high-dimesional)
- **Kernel SVM**: handles non-linear boundaries (rare in NLP pipelines)

---

#### **2.5 Logistic Regression vs SVM**

| **Feature** | **Logistic Regression** | **SVM** |
|-------------|-------------------------|---------|
| Output | Probabilities | Class (distance based) |
| Goal | Minimize error | maximize margin |
| Interpretability | High | Lower(harder to explain, which word is making more impact for a predicted class) |
| Speed | Faster | Slower(sometimes) |

#### **Advantages**

**Logistic Regression**

**SVM**

#### **Disadvantages**

**Logistic Regression**

**SVM**

#### **When to Use**

- Logistic Regression:
    - When probabilites are needed
    - Large-scale systems
- SVM:
    - When clear margin separation is needed
    - Medium-sized datasets

---

#### **When not to use**

- Complex semantic tasks (use transformers)

---

#### **Important Comparisons**

**Naive Bayes vs Logistic Regression**
- NB: probabilistic, simple
- LR: discriminative, learns better boundaries

---

#### **Logistic Regression vs SVM**
- LR: probability based (Discriminative: models learn how to separate classes)
- SVM: margin based (Generative: models learn how each class generates data)

# =========================================
# 4. IMPLEMENTATION
# =========================================

#### **4.1 Setup**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

---

#### **4.2 Sample Data**

```python
texts = [
    "Win money now",
    "Limited time offer",
    "Meeting at 5pm",
    "Project deadline tomorrow"
]

labels = ["spam", "spam", "ham", "ham"]
```

---

#### **4.3 Train-Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.5, random_state=42)
```

---

#### **4.4 Logistic Regression Pipeline**

```python
lr_pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("lr", LogisticRegression())
])

lr_pipeline.fit(X_train, y_train)
lr_preds = lr_pipeline.predict(X_test)

print("Logistic Regression:\n", classification_report(y_test, lr_preds))
```

---

#### **4.5 SVM Pipeline**

```python
svm_pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("svm", LinearSVC())
])

svm_pipeline.fit(X_train, y_train)

svm_preds = svm_pipeline.predict(X_test)

print("SVM:\n",classification_report(y_test, svm_preds))
```

---

#### **4.6 Prediction**

```python
print(lr_pipeline.predict(["Free offer just for you"]))
print(svm_pipeline.predict(["Meeting tomorrow"]))
```

---

#### **4.7 Variation: Using Embeddings**

```python
from sentence_transformers import SentenceTransformer
from sklearn.linear_model import LogisticRegression

model = SentenceTransformer("all-MiniLM-L6-v2")
X_embed = model.encode(texts)

clf = LogisticRegression()
clf.fit(X_embed, labels)
```

```python
y_embed = model.encode(["win free money", "meeting tomorrow"])

print(clf.predict(y_embed))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
texts = [
    "Stock market hits record high",
    "Investors worry about inflation rise",
    "Cryptocurrency prices drop suddenly",
    "New smartphone launch by major company",
    "Breakthrough in artificial intelligence research",
    "Cybersecurity threats increasing globally",
    "Football team wins championship",
    "Olympic games preparation begins",
    "Star player scores winning goal",
    "Bank announces new loan policies",
    "Tech company unveils new AI chip",
    "Basketball league finals start tonight",
    "Government releases economic growth report",
    "New software update improves performance",
    "Tennis player wins grand slam title"
]

labels = [
    "finance",
    "finance",
    "finance",
    "tech",
    "tech",
    "tech",
    "sport",
    "sport",
    "sport",
    "finance",
    "tech",
    "sport",
    "finance",
    "tech",
    "sport"
]

X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.5, random_state=42)

lr_pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("lr", LogisticRegression())
])

svm_pipeline = Pipeline([
    ("tfidf",TfidfVectorizer()),
    ("svm", LinearSVC())
])

lr_pipeline.fit(X_train, y_train)
svm_pipeline.fit(X_train, y_train)

lr_preds = lr_pipeline.predict(X_test)
svm_preds = svm_pipeline.predict(X_test)
```

```python
print("Logistic Regression:\n",classification_report(y_test,lr_preds))
print(lr_pipeline.predict(["tomorrow is cricket final"]))
```

```python
print("SVM:\n", classification_report(y_test, svm_preds))
print(svm_pipeline.predict(["tomorrow is cricket final"]))
```

# =========================================


1. **What is the difference between margin maximization (SVM) and probability estimation (logistic regression)**

Margin maximization in SVM focuses on finding a decision boundary that not only separates classes but does so with the largest possible margin, meaning the maximum distance between the boundary and the closest data points (support vectors). This leads to better generalization and robustness to noise. In contrast, logistic regression focuses on estimating the probability that a given input belongs to a particular class by applying a sigmoid function to a linear combination of features. Instead of maximizing separation, it optimizes the likelihood of correct classification and provides probabilistic outputs. In summary, SVM is a margin-based classifier that emphasizes geometric separation, while logistic regression is probabilistic model that outputs class probabilities.

---

2. **Why do linear models work well for high dimesional text data?**

Linear models work well for high dimesionaly text because text representation such as TF-IDF create very large feature spaces where each word acts as a separate dimension. In such spaces, classes often become linearly separable because different categories are characterized by distinct sets of words. Additionally text data is sparse, meaning most features are zero, which linear model can handle efficiently. Since these models learn weights for feature independently, then can effectively capture the contribution of individual words without needing complex interactions. This makes linear models both computationally efficient and surprisingly powerful for many NLP tasks.
 
---

3. **When would you prefer SVM over Logistic Regression?**

SVM is preferred over logistic regression when the focus is on maximizing generalization performance, especially in high dimensional settings with limited data. It is particularly effective when classes are separable with clear margin, as it emphasizes boundary robustness rather than probability estimation. While kernel SVM can handle non linear data, in NLP tasks linear SVM is usually preferred because text data is already high dimensional and often linearly separable. Compared to logistic regression, SVM may perform better in cases where decision boundary needs to be more robust to outliers, although it does not naturally provide probability outputs.
