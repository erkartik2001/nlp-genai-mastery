# =========================================
# 1. INTUITION
# =========================================

Imagine you're filtering emails into **spam** vs **not spam** using a checklist of signals:

- "free", "offer", "win" -> push toward spam
- "meeting", "project" -> push toward not spam

Each word contribute a **vote with some weight**, and you draw a **line** that separates spam from not spam.

- **Logistic Regression**: learns that line by combining weighted signals and outputs a **probability**
- **SVM**: draws a line that **maximizes the gap (margin)** between classes to make the separation more robust.

#### **What problem does this solve?**

- Classify text into categories using vector features (TF-IDF/embeddings)
- Learn a **decision boundary** that separates classes

#### **Why was this needed?**

- Naive Bayes ignore feature interactions and can be biased by independence assumption
- Needed models that:
    - Learn **optimal separating boundaries**
    - Work well in **high dimensional sparse spaces (text)**

Key Idea: 
**Turn text into vectors -> learn a boundary that separates classes**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Logistic Regression (for text)**

**Definition**

A linear model that predicts the **probability** of a class using a weight sum of features followed by a sigmoid function.

---

**Intuition**

- Each word has weight
- Sum of (word * weight) -> score
- Convert score -> probability
- Threshold (e.g 0.5) -> class

---

**Decision Boundary**

- A **linear boundary (hyperplane)** in feature space
- Separate classes based on weighted combination of features

---

**Step-by-Step**

1. Text -> TF-IDF / Embeddings
2. Compute linear score: wx+b
3. Apply sigmoid -> probability
4. Classify based on threshold

---

#### **2.2 Decision Boundary Intuition**

- In 2D a line
- In higher dimensions a hyperplane
- Points on one side -> class A
- Other side -> class B

Better weights -> better separation

---

#### **2.3 Support Vector Machine (SVM)**

**Definition**

A model that finds the **maximum margin hyperplane** that best separates classes.

---

**Intuition**

- Don't just separate classes
- Separate them with the **widest possible gap**

---

**Key Concepts**

- **Support vectors**: closest points to the boundary
- **Margin**: distance between boundary and closest points
- **Objective**: maximize margin

---

#### **2.4 Linear vs Kernel SVM**

- **Linear SVM**: works well for text (high-dimesional)
- **Kernel SVM**: handles non-linear boundaries (rare in NLP pipelines)

---

#### **2.5 Logistic Regression vs SVM**

| **Feature** | **Logistic Regression** | **SVM** |
|-------------|-------------------------|---------|
| Output | Probabilities | Class (distance based) |
| Goal | Minimize error | maximize margin |
| Interpretability | High | Lower(harder to explain, which word is making more impact for a predicted class) |
| Speed | Faster | Slower(sometimes) |

# =========================================
# 3. INTERVIEW NOTES
# =========================================

#### **Key Points**

- Both are **linear classifiers** commonly used in NLP
- Work well with **TF-IDF features**
- Handle **high-dimensional sparse data** effectively

---

#### **Advantages**

**Linear Regression**
- Output probabilities
- Easy to interpret
- Fast and scalable

**SVM**
- Strong generalization
- Robust to overfitting
- Works well in high dimensions

---

#### **Disadvantages**

**Logistic Regression**
- Can overfit without regularization
- Linear assumption

**SVM**
- No direct probabilites (needs calibaration)
- Computationally expensive

---

#### **When to Use**

- Logistic Regression:
    - When probabilites are needed
    - Large-scale systems
- SVM:
    - When clear margin separation is needed
    - Medium-sized datasets

---

#### **When not to use**

- Complex semantic tasks (use transformers)
- Deep contextual understanding problems

---

#### **Important Comparisons**

**Naive Bayes vs Logistic Regression**
- NB: probabilistic, simple
- LR: discriminative, learns better boundaries

---

#### **Logistic Regression vs SVM**
- LR: probability based (Discriminative: models learn how to separate classes)
- SVM: margin based (Generative: models learn how each class generates data)
