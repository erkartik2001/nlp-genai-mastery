# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Cosine Similarity**

**Definition**
Mesaures the **angle between two vectors**, not their magnitude

---

**Intuition**
- Same direction -> highly similar
- Opposite direction -> dissimilar
- Orthogonal -> unrelated

---

**Key Insight**
- Focused on orientation, not size
- Perfect for text (where length varies)

---

**Range**
- 1 -> identical
- 0 -> no similarity
- -1 -> opposite

---

#### **2.2 Euclidean Distance**

**Definition**

Mesaures the **straight line distance** between 2 vectors

---

**Intuition**

- Closer points -> more similar
- Far apart -> less similar

---

**Key Insight**
- Sensitive to magnitude (vector length)
- Not ideal when document length differ

---

#### **2.3 Cosine vs Euclidean**

| **Feature** | **Cosine Similarity** | **Euclidean Distance** |
|-------------|-----------------------|------------------------|
| Focus | Direction | Distance |
| Scale sensitivity | No | Yes |
| Best for text | Yes | Limited |

---

#### **2.4 Document Similarity Systems**

**Definition**

Systems that:
- Convert text -> vectors
- Compare vectors -> similarity score
- Rank results

---

#### **2.5 Step-by-Step Working**

1. Preprocess text
2. Convert to vectors (TF-IDF, embeddings)
3. Compute similarity (cosine/distance)
4. Rank results
5. Return most similar documents

---

# =========================================
# 3. INTERVIEW NOTES
# =========================================

#### **Key Points**

- Similarity is computed on vector representations
- Cosine similarity is most common used in NLP
- Euclidean distance is less suitable for high-dimensional sparse data

---

#### **Advantages**

**Cosine**
- Scale invariant
- Works well for sparse vector
- Ideal for TF-IDF

**Euclidean**
- Simple 
- Useful in low-dimensional dense spaces

---

#### **Disadvantages**

**Cosine**
- Ignores magnitude differences

**Euclidean**
- Sensitive to scaling
- Poor for high-dimensional sparse data

---

#### **When NOT to Use**

- Euclidean for sparse TF-IDF
- Cosine when magnitude matters

---

#### **Important Comparisons**

**TF-IDF + Cosine vs Embeddings + Cosine**

| **Feature** | **TF-IDF** | **Embeddings** |
| Semantics | No | Yes |
| Use case | Keyword match | Semantic search |

# =========================================
# 4. IMPLEMENTATION
# =========================================

#### **4.1 Setup**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
```

---

#### **4.2 Sample Data**

```python
docs = [
    "I love NLP",
    "I enjoy NLP",
    "I hate cooking"
]
```

---

#### **4.3 TF-IDF Vectorizer**

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)
```

---

#### **4.4 Cosine Similarity**

```python
cos_sim = cosine_similarity(X)
print("Cosine Similarity:\n", cos_sim) # higher more similar
```

---

#### **4.5 Euclidean Distance**

```python
euc_dist = euclidean_distances(X)
print("Euclidean Distance:\n",euc_dist) # less distance more similar
```

---

#### **4.6 Document Ranking**

```python
def rank_docs(query, docs):
    vectorizer = TfidfVectorizer()
    X = vectorizer.fit_transform([query]+docs)

    sim = cosine_similarity(X[0:1], X[1:]).flatten()
    ranked = sorted(zip(docs, sim), key = lambda x: x[1], reverse = True)

    return ranked

print(rank_docs("I love NLP",docs))
```

---

#### **4.7 Variation: Using Embeddings**

```python
# from sentence_transformers import SentenceTransformer
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(docs)


cos_sim = cosine_similarity(embeddings)
print(cos_sim)
```

# 8. SELF-TEST QUESTIONS
# =========================================

1. **Why do cosine similarity preferred over Euclidean distance in NLP?**

Cosine similarity is preferred over euclidean distance in NLP because it measures the orientation (angle) between vectors rather than their magnitude. In text representation such as TF-IDF or embeddings, document length and word frequency can significantly affect vector magnitude, which can distort euclidean distance. Cosine similarity removes this issue by only focusing on the relative distribution of terms, making it robust to difference in document lenght. Additionally, NLP vectors are often high dimensional and sparse, where euclidean distance becomes less meaningful due to the curse of dimenisonality. As a result, cosine similarity provides more reliable similarity comparisons in most NLP applications.

---

2. **How does vector representation affect similarity results?**

Vector representation directly determines the quality and type of similarity results because similarity matrix operate entirely on these vectors. Different representation captures different aspects of text: for example, Bag of Words and TF-IDF focus on word frequency and keyword overlap, while embeddings capture semantic meaning and contextual relationships. If the represenation lacks semantic information, similarity results will reflect surface level matching rather than true meaning. Conversely, richer representations like embeddings allow the system to identify similarity even when different words convey similar meanings. Therefore, the choice of vector represenation is often more important than the similarity metric itself, as it defines that "similarity" actually means in the system. 

---

3. **Why does TF-IDF + cosine fail for semantic similarity?**

TF-IDF combined with cosine similarity fails on semantic similarity because it represents texts as independent weighted words without capturing context, meaning or relationships between them. It relies purely on terms frequency and inverse document frequency, which emphasize keyword overlap rather than understanding. As a result, two sentences with similar meanings but different words (e.g., "car", "automobile") may appear dissimilar, while sentences sharing common words but different intent may appear similar. This limitation makes TF-IDF effective for keyword based retrieval but inadequate for semantic tasks, where embeddings and transformer based models are more appropriate.
