Imagine reading this sentence: **"Apple hired john in california"**

As humans, we naturally understand:
- Apple -> company
- john ->  person name
- california -> location

- hired -> verb
- john -> noun

Sequence labeling teaches machines: **"Assign a label to every token in sequence."**

---

#### **What problem does this solve?**
- Understanding structure and meaning inside text
- Identifying:
    - Grammar roles
    - Entities
    - Relationships

---

#### **Why was this needed?**

Many NLP tasks require token-level understanding:

- Chatbots
- Search systems
- Voice assistants
- Information extraction

Without sequence labeling:
- Machine know words
- But not their role or meaning in context

Key Idea: **Input sequence of words -> output sequence of labels**

#### **2.4 Hidden Markov Models (HMM)**

**Definition**

A probabilistic sequence model that predicts hidden states from observed words. Has strong independence assumptions.

---

**Intuition**

Words are visible:

**"dog barks loudly"**

But POS tags are hidden:

**NOUN VERB ADVERB**

HMM tries to infer:

**"What hidden label sequence most likely generated these words?"**

---

**HMM assumes:**
- Current tag depends on previous tag (P(yi | yi-1)), not entire sequence
- Current words depends only on current tag (P(xi | yi)), not neighbour words

---

**Key Components**

| **Component** | **Meaning** |
|---------------|-------------|
| Hidden States | Labels (POS Tags) |
| Observations | Words |
| Transition Probability | Probability of next tag |
| Emission Probability | Probability of word given tag |

---

**Step-by-Step**

1. Observe word sequence
2. Estimate transition probability
3. Estimate emission probability
4. Predict most likely label sequence

---

**HMM computes:**

P(tags,words) using:

P(W,T)=∏P(ti∣ti−1)P(wi∣ti)

Meaning of formula

For each position:

(previous tag -> current tag) × (current tag -> current word)

---

#### **2.5 CRF (Conditional Random Field)**

**Definition**

A probabilistic sequence labeling model that predicts labels considering the full sequence context.

---

**Intuition**

HMM predicts using local probabilities.

CRF improves this by considering.

- Neighboring labels
- Entire sequence consistency

---

**Example**

"New York"

CRF understands:

- If "New" is LOCATION part,
- "York" is likely continues LOCATION

---

**CRF assumes:**
- Use any useful features from entire sequence

---

**Why CRF Better Than HMM?**

- Uses richer features
- Handles dependencies better
- Avoids unrealistic independence assumptions
- Language depends heavily on context, CRF removed HMM's unrealistic assumptions

---

#### **2.6 HMM vs CRF**

| **Feature** | **HMM** | **CRF** |
|-------------|---------|---------|
| Intuition | How could these words and tags can be generated | Given this whole sentence, Which labels make most sense? |
| Type | Generative | Discriminative |
| Assumption | Strong independence | Flexible |
| Performance | Moderate | Better (context) |
| Learns | Probabilities (P(word,tag), P(tag,previous tag)) | Feature weights |
| Models | Joint probability P(X,Y) | Conditional Probability P(Y|X) |

---

**Joint vs Conditional Probability**

Joint Probability (Generative)
- P(X,Y)
- Probability of input AND label together
- P(Sentence,tags)

Conditional Probability (Discriminative)
- P(Y|X)
- Probability of label GIVEN input
- P(Spam | message)

#### **Project: Resume Entity Extractor**

**Input**

---

**Expected Ouput**

- John -> PERSON
- Microsoft -> ORGANIZATION
- California -> LOCATION
- 2020 -> DATE

1. Tokenize text

1. **Why is CRF generally preferred over HMM for sequence labeling tasks?** 

CRF or conditional random field is generally preferred over HMM for sequence labeling tasks because it models the entire label sequence conditionally while incorporating rich contextual features from the input text. HMM relies on strong independence assumptions, such as assuming current state depends only on the previous state and that observations are generated independently. These assumptions are often unrealistic for natural language. In contrast, CRF can use multiple contextual features simultaneously, including neighboring words prefixes, suffixes, capitalization and surrounding tags, which allows it to capture sequence dependencies more effectively. Since language is high context dependent, CRF produces more accurate and consistent label predictionsd than HMM in tasks such as POS and NER.

---

2. **What is the difference between POS tagging and Named Entity Recognition?**

POS and NER are both are sequence labeling tasks, but they focus on different types of information. POS tagging assigns grammatical labels to words, such as noun, verb, adjective or adverb helping the systems understand the syntactic structure of sentence. For example, in the sentence "kartik ran quickly" POS tagging identifies kartik as a noun, ran as verb and quickly as adverb. In contrast, NER identifies and categorizes real world entities such as person names, locations, organizations, and dates. For example in "kartik is living in indore since 2001" NER identifies, kartik as person, indore as location, 2001 and a date. In summary, POS tagging focuses on grammatical roles, while NER focuses on extracting meaningful entities from text. Bother help convert unstructured text into structured information for downstream in NLP systems.

---

3. **Why are transformers replacing classical sequence models like HMM and CRF?**

Transformers are replacing classical sequence models like HMM and CRF because they can capture long range contextual and semantic relationships much more effectively. HMM and CMM rely heavily on handcrafted features and limited contextual assumptions, whereas transformers learn contextual representations automatically using self attention mechanisms. This allows transformers to understand relationships between words across an entire sequence rather than depending mainly on local transitions. Additionally transformers are pretrained on massive corpora, enabling them to learn grammar, semantics, entities and sequence dependencies simultaneously. As a result transformer based model achieve significantly better performance on tasks such as POS tagging, NER, translation, and question answering, while also reducing the need of manual feature engineering.
