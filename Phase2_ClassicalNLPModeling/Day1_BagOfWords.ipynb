# =========================================
# 1. INTUITION
# =========================================

Imagine you have a box of words, and you only care about **which words appear and how many times**, not the order in which they appear.

**Example:**

- "I love NLP"
- "NLP love I"

For humans, these feel different.
For a Bag of Words model -> **they are exactly the same**.

#### **What problem does this solve?**

- Machines need **numerical representation of text**
- Raw text cannot be directly used in ML models
- BoW converts text -> **fixed length vectors**

#### **Why was this needed?**

- Early ML models (Logistic Regression, Naive Bayes) require numbers
- Need a simple way to:
    - Represent text
    - Capture word importance
- BoW was one of the **first and simplest solutions**

Key Idea:
**Ignore order -> focus only on word presence and frequency**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Bag of Words (Definition)**

A text representation technique where a document is represented as a **vector of word counts** based on a fixed vocabulary.

---

#### **2.2 Representation**

**Step 1: Build Vocabulary**

Corpus: ["I love NLP", "I love AI"]

Vocabulary: ["I", "love", "NLP", "AI"]

---

**Step 2: Convert to Vectors**

"I love NLP" -> [1, 1, 1, 0]
"I love AI" -> [1, 1, 0, 1]

Each index corresponds to a word in vocabulary.

---

#### **2.3 Sparse Vectors**

**Definition**
Vectors where most values are **zero**

**Example:** [0, 0, 1, 0, 0, 1, 0]

---

#### **Why Sparse?**
- Vocabulary is large
- Each document uses only a few words

---

#### **Impact**
- Memory inefficiency (if not optimized)
- Requires spare matrix representation

---

#### **2.4 Step-by-Step Working**

1. Collect corpus
2. Tokenize text
3. Build vocabulary
4. Assign index to each word
5. Count word frequency per document
6. Create vector representation

---

#### **2.5 Limitations**

1. **No Word Order**
- "I love NLP" = "NLP love I"

---

2. **No Context Understanding**
- Cannot distinguish:
    - "good" vs "not good"

---

3. **High Dimensionality**
- Large vocabulary - large vectors

---

4. **Sparsity**
- Most values are zero

---

5. **OOV Problem**
- New words are ignored

# 4. IMPLEMENTATION

#### **4.1 Using sklearn CountVectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love AI",
    "I love NLP"
]

# corpus = [
#     "I love AI NLP",
#     "I love NLP"
# ]


vectorizer = CountVectorizer() # converts to lowercase, removes very short tokens (single letters)
x = vectorizer.fit_transform(corpus)

print(x) # Sparse matrix, efficient storage


print("Vocabulary:", vectorizer.get_feature_names_out()) 

# Only stores non-zero values + their positions, indexing is memory efficient (0 row,1 column) = 1, Dense [1,1,0]
# We note only the index which is non zero
# A sparse matrix stores only the positions (indices) where values are non-zero + their values.
# Bag of Words vector representation of both sentences

print("Vectors:\n", x.toarray()) # Dense matrix (human readable) full representation, not stored internally only generated
```

---

#### **4.2 Manual Implementation**

```python
from collections import Counter

def build_vocab(corpus):
    vocab = set()

    for text in corpus:
        # print(text.lower().split())
        # print(vocab)
        vocab.update(text.lower().split())
    return sorted(vocab)

def vectorize(text, vocab):
    word_count = Counter(text.lower().split())
    return [word_count[word] for word in vocab]

corpus = ["I love NLP", "I love AI"]

vocab = build_vocab(corpus)

vectors = [vectorize(text, vocab) for text in corpus]

print("Vocabulary: ",vocab)
print("Vectors: ",vectors)
```

# 6. MINI PROJECT / TASK
# =========================================


#### **Project: Spam Classifier using Bow**

**Input:**

[
    "Win money now",
    "Hello friend",
    "Claim your prize now"
]

--- 

#### **Expected Output**
- BoW vectors
- Classification (spam / not spam)

---

#### **Steps**
1. Preprocess text
2. Convert to BoW vectors
3. Train classifier (Naive Bayes)
4. Evaluate performance

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

texts = [
    "Win money now",
    "Hello friend",
    "Claim your prize now"
]

labels = [
    "spam",
    "ham",
    "spam"
]

# convert text to BoW vectors
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(texts)

# Train naive bayes model
model = MultinomialNB()
model.fit(x, labels)

# test sentences
test_sentences = [
    "win free money",
    "hi brother"
]

# convert test sentences to vectors 
x_test = vectorizer.transform(test_sentences)

# predict
predictions = model.predict(x_test)

print(predictions)
```

# 8. SELF-TEST QUESTIONS
# =========================================

1. **Why does Bag of Words fail to capture context?**

Bag or Words fails to capture context because it represents text purely based on word frequency while completely ignoring the order and relationships between words. As a result, sentences with different meanings but identical word counts, such as "kartik love NLP" and "NLP love kartik" are treated as the same representation. Beyond word order, BoW also lacks any understanding of semantics, meaning it cannot distinguish nuances like negation ("good" vs "not good") or contextual usage of words. Since it does not model dependencies or relationships between tokens, it cannot capture meaning beyound isolated word presence, which severly limits its effectiveness for tasks requiring contextual understanding.

---

2. **What is a sparse vector and why does BoW produce it?**

A sparse vector is a high dimensional vector in which most of the values are zero, with only a small number of non-zero entries. Bag of Words produces sparse vectors because the vocab constructed from a corpus is typically very large, while any individual document contain only a small subset of those words. As a result, most positions in the vectors corresponds to absent words are filled with zeros. This leads to inefficeincies in memory and computation if stores as dense vectors. To address this BoW representations are typically stored using sparse matrix formats, which only record the indices and values of non zero elements, making them more memory efficient and computationally practical.
