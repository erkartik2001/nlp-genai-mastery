#### **4.1 Spam Classifier**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

texts = [
    "Win a free iPhone now",
    "Limited time offer claim your prize",
    "Get cash rewards instantly",
    "Congratulations you have won a lottery",
    "Exclusive deal just for you click now",
    
    "Let's schedule a meeting tomorrow",
    "Project deadline is next week",
    "Can you review the document",
    "Lunch plan at 1pm",
    "Please call me when you are free"
]

labels = [
    "spam",
    "spam",
    "spam",
    "spam",
    "spam",
    
    "ham",
    "ham",
    "ham",
    "ham",
    "ham"
]

X_train, X_test, y_train, y_test = train_test_split(texts, labels, test_size=0.25, random_state=42)

pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", MultinomialNB())
])

pipeline.fit(X_train, y_train)

preds = pipeline.predict(X_test)

print(classification_report(y_test, preds))
```

#### **4.2 Document Similarity System**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "Machine learning is powerful",
    "Deep learning is part of AI",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

print("Vocab: ",end=" ")
print(vectorizer.get_feature_names_out(), "\n\n")


print(X,"\n\n")

print("Amplified sparse matrix, for each doc, each document becomes point in vector space")
print(X.toarray(), "\n\n") 

x = ["machine learning"]

y = vectorizer.transform(x) ## Query should be in same vocabulary space

similarity_matrix = cosine_similarity(y,X)

print("Query transformed ",y.toarray())
print("Most similar document",similarity_matrix)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

texts = [
    "I want a refund",
    "Track my package",
    "Cancel my order",
    "Where is my delivery",
    "I need my money back",
    "Stop my shipment",
    "Track my parcel status",
    "Cancel the purchase",
    "Refund for damaged item",
    "My order tracking is not updating",
    "Please cancel this item",
    "I have not received my package"
]

labels = [
    "refund",
    "tracking",
    "cancel",
    "tracking",
    "refund",
    "cancel",
    "tracking",
    "cancel",
    "refund",
    "tracking",
    "cancel",
    "tracking"
]


pipeline = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", LogisticRegression())
])

pipeline.fit(texts, labels)

queries = ["where is my package", "cancel his appointment", "where is my refund"]

for query in queries:
    print(query,": ", pipeline.predict([query]))
```

# =========================================


1. **Why does TF-IDF work well for spam classification and document retrieval tasks?**

TF-IDF works well for spam classification and document retrieval because it highlights informative and discriminative words while reducing the influence of extremely common words. Unlike BoW, which treats all the words equally on the basis of their frequency. TF-IDF assigns weights using both local importance within a document (term frequency) and global rarity across the corpus (IDF). This helps emphasize keywords such as "free', "offer", or domain specific terms that are important for distinguishing classes or retrieving relevant documents. The resulting sparse high dimensional vectors work particularly well with linear classifiers and similarity based retrieval systems, making TF-IDF highly effective for keyword-oriented NLP tasks even without deep semantic understanding.

---

2. **Why is cosine similarity preferred over Euclidean distance for similarity?**

Cosine similarity is preferred over euclidean distance in NLP because it measures the orientation or angular similarity between vectors rather than their absolute magnitude. In text representation such as TF-IDF or embeddings, document length and word frequency can significantly affect vector magnitude, making euclidean distance unreliable for comparing semantic similarity. Two document with similar meanings but different lengths may appear apart using euclidean distance. Cosine similarity avoids this issue by focusing only on the directional relationship between vectors, making its more robust or high-dimensional sparse text data. This is why cosine similarity is widely used in document retrieva, search systems and semantic similarity pipelines.

---

3. **Why are modern embedding systems replacing TF-IDF based similarity systems?**

Modern embedding systems are replacing TF-IDF based similarity systems because embeddings capture semantic and contextual relationships between words and sentences, whereas TF-IDF relies mainly on keyword frequency and rarity. TF-IDF cannot understand synonyms, context or semantic similarity between different phrases with similar meanings. In contrast, embeddings generated by transformer models encode contextual meaning into dense vector representations, allowing semantically related text to appear close in vector space even if they do not share exact keywords. This significantly improves retrieval quality, semantic search, recommendation system, and RAG pipelines. As a result embedding based system provide much stronger performance for modern NLP applications that require contextual understanding rather than simple keyword matching.
