# =========================================
# 1. INTUITION
# =========================================

Imagine you're trying to teach a child to read a new language. Instead of memorizing every possible word, you teach them:

- Small meaningful pieces like "un", "happy", "ness"
- So they can understand **new words they've never seen before**

Subword tokenization works the same way:
- Instead of storing full words -> break them into **smaller reusable units**

#### **What problem does this solve?**
- **Out-of-Vocabulary (OOV) problem**
    - Traditional tokenization fails on unseen words
- Huge vocabulary size
- Poor generalization with on rare words

#### **Why was this needed?**
- LLMs are trained on massive text with:
    - Rare words
    - Typos
    - Multiple Languages
- Subword tokenization allows:
    - Smaller vocabulary
    - Better generalization
    - Efficient learning

Key Idea: **Learn smaller building blocks -> Understand infinite words**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Subword Tokenization (Definition)**
A method of splitting words into **smaller meaningful units (subwords)** instead of full words.

Example: "unhappiness" -> ["un", "happi", "ness"]

---

#### **2.2 Why Subword Instead of Word?**

| **Problem** | **Word Tokenization** | **Subword Tokenization** |
|-------------|-----------------------|--------------------------|
| Unknown words | No | Yes |
| Vocabulary Size | Huge | Controlled |
| Generalization | Poor | Strong |

---

#### **2.3 Byte Pair Encoding (BPE)**

**Idea**
Start with characters ->  iteratively merge **most frequent pairs**

---

**Steps**
1. Start with character-level tokens
2. Count most frequent adjacent pairs
3. Merge the most frequent pairs 
4. Repeat until predifined vocab size (by user) reached

---

Example:  ["low", "lower", "lowest"]

**Steps:**
- l o w -> merge "lo"
- lo w -> merge "low"
- low er -> merge "lower"

Final Tokens: ["low", "er", "est]

--- 

**Key Insight**
- Frequency based merging
- Simple and effective

---

#### **2.4 WordPiece (Used in BERT)**

**Idea**
Similar to BPE but uses **probablistic scoring**

Instead of frequency:
- Chooses merges that maximize likelihood of corpus
- In simple terms choose merges that make the corpus easier to represent using fewer tokens

What does likelihood mean here, We want tokens that:
- appear often
- AND are useful across many words

---

Example: "playing" -> ["play", "##ing""]
- "##" indicates conitnuation

---

**Key Insight**

- More optimal than BPE
- Used in transformers models like BERT
- Favors meaningful combinations, not just frequent letters
- WordPiece prefers useful building blocks, not just frequent pairs

---

#### **2.5 Sentence Piece**

**Idea**

Treat text as **raw stream (no spaces assumed)**
- Works directly on full setences
- Language independent

---

**Key Features**

- No need for pre-tokenization
- Handles multilingual text well

---

Example: "I love NLP" -> ["_I", "_love", "_NLP"]
("_" represents space)

---

#### **2.6 Unigram Language Model Tokenizer**

**Idea**

Start with large vocab -> removes token based on probability. 

---

**Steps**
1. Initialize large vocabulary
2. Compute probability of tokens
3. Remove least useful tokens, If removing a token doesn’t hurt representation much → remove it
4. Optimize likelihood

---

**Key Insight**
- Probabilistic approach
- More flexible segmentation

Key Idea: **Start with a large vocab → remove bad tokens based on probability assigned to each token (till vocab size reached)**

---

#### **2.7 Step-by-Step Flow

Input: "unbelievable"

Subword tokens: ["un", "believ", "able"]

Process:
1. Break into subwords
2. Convert into IDs
3. Feed into model

# =========================================
# 3. INTERVIEW NOTES
# =========================================

#### **Key Points**
- Subword tokenization solve OOV problem
- Balances vocabulary size and expressiveness
- Used in all modern LLMs

---

#### **Advantages**

- Handles unseen words
- Reduces vocabulary size
- Better generalization
- Efficient training

---

#### **Disadvantages**

- Can split words unnaturally
- Slightly increase squence length
- Complexity in training tokenizer

---

#### **When to Use**

- LLMs
- Transformer models
- multilingual systems

---

#### **When Not to Use**

- Very simple NLP tasks
- Small dataset with limited vocabulary

---

#### **Important Comparisons**

#### **BPE vs WordPiece vs Unigram**

| **Feature** | **BPE** | **WordPiece** | **Unigram** |
|-------------|---------|---------------|-------------|
| Approach | Frequency | Probabilistic | Probabilistic |
| Training | Greedy | Optimized | Global Optimization |
| Flexibility | Medium | High | Very High|

---

**SentencePiece vs Others**
- Works without pre-tokenization
- Better for multilingual pipelines
