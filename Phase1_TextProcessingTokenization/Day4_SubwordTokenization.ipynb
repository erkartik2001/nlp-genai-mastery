# 1. INTUITION
# =========================================

Imagine you're trying to teach a child to read a new language. Instead of memorizing every possible word, you teach them:

- Small meaningful pieces like "un", "happy", "ness"
- So they can understand **new words they've never seen before**

Subword tokenization works the same way:
- Instead of storing full words -> break them into **smaller reusable units**

#### **What problem does this solve?**
- **Out-of-Vocabulary (OOV) problem**
    - Traditional tokenization fails on unseen words
- Huge vocabulary size
- Poor generalization with on rare words

#### **Why was this needed?**
- LLMs are trained on massive text with:
    - Rare words
    - Typos
    - Multiple Languages
- Subword tokenization allows:
    - Smaller vocabulary
    - Better generalization
    - Efficient learning

Key Idea: **Learn smaller building blocks -> Understand infinite words**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Subword Tokenization (Definition)**
A method of splitting words into **smaller meaningful units (subwords)** instead of full words.

Example: "unhappiness" -> ["un", "happi", "ness"]

---

#### **2.2 Why Subword Instead of Word?**

| **Problem** | **Word Tokenization** | **Subword Tokenization** |
|-------------|-----------------------|--------------------------|
| Unknown words | No | Yes |
| Vocabulary Size | Huge | Controlled |
| Generalization | Poor | Strong |

---

#### **2.3 Byte Pair Encoding (BPE)**

**Idea**
Start with characters ->  iteratively merge **most frequent pairs**

---

**Steps**
1. Start with character-level tokens
2. Count most frequent adjacent pairs
3. Merge the most frequent pairs 
4. Repeat until predifined vocab size (by user) reached

---

Example:  ["low", "lower", "lowest"]

**Steps:**
- l o w -> merge "lo"
- lo w -> merge "low"
- low er -> merge "lower"

Final Tokens: ["low", "er", "est]

--- 

**Key Insight**
- Frequency based merging
- Simple and effective

---

#### **2.4 WordPiece (Used in BERT)**

**Idea**
Similar to BPE but uses **probablistic scoring**

Instead of frequency:
- Chooses merges that maximize likelihood of corpus
- In simple terms choose merges that make the corpus easier to represent using fewer tokens

What does likelihood mean here, We want tokens that:
- appear often
- AND are useful across many words

---

Example: "playing" -> ["play", "##ing""]
- "##" indicates conitnuation

---

**Key Insight**

- More optimal than BPE
- Used in transformers models like BERT
- Favors meaningful combinations, not just frequent letters
- WordPiece prefers useful building blocks, not just frequent pairs

---

#### **2.5 Sentence Piece (not an algorithm)**

**Idea**

Treat text as **raw stream (no spaces assumed)**
- Works directly on full setences
- Language independent

---

**Key Features**

- No need for pre-tokenization
- Handles multilingual text well

---

Example: "I love NLP" -> ["_I", "_love", "_NLP"]
("_" represents space)

---

#### **2.6 Unigram Language Model Tokenizer**

**Idea**

Start with large vocab -> removes token based on probability. 

---

**Steps**
1. Initialize large vocabulary
2. Compute probability of tokens
3. Remove least useful tokens, If removing a token doesn’t hurt representation much → remove it
4. Optimize likelihood

---

**Key Insight**
- Probabilistic approach
- More flexible segmentation

Key Idea: **Start with a large vocab → remove bad tokens based on probability assigned to each token (till vocab size reached)**

---

#### **2.7 Step-by-Step Flow

Input: "unbelievable"

Subword tokens: ["un", "believ", "able"]

Process:
1. Break into subwords
2. Convert into IDs
3. Feed into model

#### **4.2 BPE Tokenizer (HugginingFace)**

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace

tokenizer = Tokenizer(BPE())
tokenizer.pre_tokenizer = Whitespace()

trainer = BpeTrainer(vocab_size=1000)

files = ["sample.txt"]
tokenizer.train(files, trainer)

output = tokenizer.encode("I love NLP")
print(output.tokens)
print(tokenizer.get_vocab())
print(tokenizer.get_vocab_size())

""" 
After subword tokenization, each token is mapped to a number or token id, then embeddings will be generated 
of particular fixed size vector (768, 1024) dim. Then positional encodings will be added, because order matters.
Token embedding + Position embedding, then passed to the model.

#Raw text -> Tokenizer (BPE) -> Tokens -> Token IDs -> Embeddings -> Transformer (LLM) -> Output (prediction) """
```

#### **4.3 WordPiece Tokenizer**

```python
from transformers import BertTokenizer

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

tokens = tokenizer.tokenize("unbelievable") # Exists in bert vocab no need to tokenize
print(tokens)

tokens = tokenizer.tokenize("unbelievablen") # Does not exists in vocab
print(tokens)

print(tokenizer.tokenize("supercalifragilisticexpialidocious")) # Not present in bert vocab
```

#### **4.4 SentencePiece**

```python
import sentencepiece as spm

# Unigram subword tokenization
spm.SentencePieceTrainer.train( 
    input= "sample.txt", 
    model_prefix='m',
    vocab_size=400 # Vocab size should be small for small corpus
)

sp = spm.SentencePieceProcessor()
sp.load('m.model')

print(sp.encode("I love NLP", out_type=str))
```

1. **Why do Subword tokenizers perform better than word tokenizer in LLMs?**

Subword tokenizers perform better than word tokenizers because they effectively solve the out-of-vocabulary (OOV) problem while improving generalization across rare and unseen words. Instead of treating each word as a separate unit, subword tokenization breaks words into smaller, reusable components such as prefixes, roots, and suffixes (e.g., “stronger”, “wider”, “lowest” → “er”, “est”). These subunits are learned based on frequency or probabilistic patterns across the corpus, allowing the model to construct and understand new words it has never explicitly seen during training. This significantly reduces vocabulary size while maintaining expressive power. Additionally, subword tokenization works well across multiple languages and diverse text distributions, making training more efficient and scalable for modern LLMs.

---

2. **What is the key difference between BPE and WordPiece?**

The key difference between Byte Pair Encoding (BPE) and WordPiece lies in how they decide which subword units to merge during training. BPE follows a frequency-based approach, where it iteratively merges the most frequent adjacent character or subword pairs in the corpus until a desired vocabulary size is reached. In contrast, WordPiece uses a probabilistic approach, selecting merges that maximize the likelihood of the training data under a language model. This makes WordPiece more optimized in terms of representation quality, as it considers the overall contribution of a merge to the corpus rather than just raw frequency. As a result, WordPiece generally produces more meaningful and efficient tokenizations compared to BPE.

---

3. **Why is sentence piece useful for multilingual systems?**
 
SentencePiece is particularly useful for multilingual systems because it operates directly on raw text without relying on whitespace-based tokenization, making it language-independent. Many languages such as Chinese and Japanese do not use spaces to separate words, which makes traditional tokenization methods ineffective. SentencePiece treats the input as a continuous stream of characters and learns subword units directly from this raw data, allowing it to handle diverse linguistic structures uniformly. This flexibility enables it to work seamlessly across multiple languages, scripts, and writing styles, making it a strong choice for training multilingual LLMs.
