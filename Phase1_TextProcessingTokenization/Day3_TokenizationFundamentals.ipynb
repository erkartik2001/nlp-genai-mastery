# =========================================
# 1. INTUITION
# =========================================

Imagine you're trying to teach a machine to read a paragraph.
To you, a sentence is natural, but for a machine its just **long string of characters.**

Tokenization is like **breaking a paragraph into meaningful lego blocks:**

- Sentences - paragraph splits into sentences
- Words - sentences split into words
- Characters - words split into letters

#### **What problem does this solve?**

- Machine cannot undestand raw continuous text
- They need **discrete units (tokens)** to process language
- Tokenization converts text into **processable units**

#### **Why was this needed?**

- All NLP models (from classical to LLMs) works on tokens.
- Without tokenization:
    - No embeddings
    - No training
    - No understanding

Tokenization is the **first most critical step in any NLP pipeline**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Tokenization (definition)**
The process of splitting text into smaller units called **tokens**

Tokens can be:
- Words
- Sentences
- Characters
- Subwords (in modern LLMs)

---

#### **2.2 Type of Tokenization**

#### **1. Word Tokenization**
Splits text into words

**Example:**
- "i love NLP" -> ["i", "love", "NLP"]

Most common, used in classical NLP

---

#### **2. Sentence Tokenization**
Splits paragraph into sentences

**Example:**
- "I love NLP. It is amazing!" -> ["I love NLP", "It is amazing"]

**Useful in:**
- Summarization 
- Question answering
- Document processing

---

#### **3. Character Tokenization**
Splits text into characters

**Example:**
- "cat" -> ["c", "a", "t"]

**Used in:**
- Language modeling in early models (Language Modeling is a technique used in natural language processing that involves predicting the next word in a sentence or sequence of words based on the context and previous words)
- Handling unknown words

But loses semantic meanining

---

#### **2.3 Tokenization Boundary Problems**

#### **What is it?**
Deciding **where to split tokens correctly**

---

#### **Common Issues**

**1. Contradictions**
- "don't" -> ["do", "n't"] or ["don't"]?

**2. Punctuation**
- "hello!" -> ["hello","!"] or ["hello!"]?

**3. Hyphenated Words**
- "state-of-the-art" -> one token or multiple?

**4. Named Entities**
- "New York" -> one token or two?

**5. Language without spaces**
- Chinese, Japanese -> no clear word boundaries

Tokenization is not trivial, it involves **linguistic decisions**

---

#### **2.4 NLTK vs spaCy Tokenization**

**NLTK**
- Rule based
- Simpler and flexible 
- Good for learning and experimentation

**Example:**
from nltk.tokenize import word_tokenize


---

**spaCy**
- Industrial-grade tokenizer
- Uses statistical + rule based approach
- Handles edge cases better

**Example:**
import spacy
nlp = spacy.load("en_core_web_sm")
