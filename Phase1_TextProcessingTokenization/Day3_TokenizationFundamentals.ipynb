# =========================================
# 1. INTUITION
# =========================================

Imagine you're trying to teach a machine to read a paragraph.
To you, a sentence is natural, but for a machine its just **long string of characters.**

Tokenization is like **breaking a paragraph into meaningful lego blocks:**

- Sentences - paragraph splits into sentences
- Words - sentences split into words
- Characters - words split into letters

#### **What problem does this solve?**

- Machine cannot undestand raw continuous text
- They need **discrete units (tokens)** to process language
- Tokenization converts text into **processable units**

#### **Why was this needed?**

- All NLP models (from classical to LLMs) works on tokens.
- Without tokenization:
    - No embeddings
    - No training
    - No understanding

Tokenization is the **first most critical step in any NLP pipeline**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Tokenization (definition)**
The process of splitting text into smaller units called **tokens**

Tokens can be:
- Words
- Sentences
- Characters
- Subwords (in modern LLMs)

---

#### **2.2 Type of Tokenization**

#### **1. Word Tokenization**
Splits text into words

**Example:**
- "i love NLP" -> ["i", "love", "NLP"]

Most common, used in classical NLP

---

#### **2. Sentence Tokenization**
Splits paragraph into sentences

**Example:**
- "I love NLP. It is amazing!" -> ["I love NLP", "It is amazing"]

**Useful in:**
- Summarization 
- Question answering
- Document processing

---

#### **3. Character Tokenization**
Splits text into characters

**Example:**
- "cat" -> ["c", "a", "t"]

**Used in:**
- Language modeling in early models (Language Modeling is a technique used in natural language processing that involves predicting the next word in a sentence or sequence of words based on the context and previous words)
- Handling unknown words

But loses semantic meanining

---

#### **2.3 Tokenization Boundary Problems**

#### **What is it?**
Deciding **where to split tokens correctly**

---

#### **Common Issues**

**1. Contradictions**
- "don't" -> ["do", "n't"] or ["don't"]?

**2. Punctuation**
- "hello!" -> ["hello","!"] or ["hello!"]?

**3. Hyphenated Words**
- "state-of-the-art" -> one token or multiple?

**4. Named Entities**
- "New York" -> one token or two?

**5. Language without spaces**
- Chinese, Japanese -> no clear word boundaries

Tokenization is not trivial, it involves **linguistic decisions**

---

#### **2.4 NLTK vs spaCy Tokenization**

**NLTK**
- Rule based
- Simpler and flexible 
- Good for learning and experimentation

**Example:**
from nltk.tokenize import word_tokenize


---

**spaCy**
- Industrial-grade tokenizer
- Uses statistical + rule based approach
- Handles edge cases better

**Example:**
import spacy
nlp = spacy.load("en_core_web_sm")

---

#### **2.5 Step by step working**

Input: "I don't love NLP!!! It's hard."

Steps:
1. Sentence tokenization
2. Word tokenization
3. Handle punctuation
4. Handle contractions

Output (spaCy-like):

#### **4.1 Setup**

```python
import nltk
import spacy

nltk.download("punkt")
```

---

#### **4.2 Word Tokenization (NLTK)**

```python
from nltk.tokenize import word_tokenize

text = "I love NLP!"

tokens = word_tokenize(text)
print(tokens)
```

---

#### **4.3 Sentence Tokenization (NLTK)**

```python
from nltk.tokenize import sent_tokenize

text = "I love NLP. It is amazing!"

sentence_tokens = sent_tokenize(text)
print(sentence_tokens)
```

---

#### **4.4 Character Tokenization**

```python
text = "NLP"

char_tokens = list(text)
print(char_tokens)
```

---

#### **4.5 spaCy Tokenization**

```python
nlp = spacy.load("en_core_web_sm")

text = "I don't love NLP!!! It's hard."

doc = nlp(text)

tokens = [tokens.text for tokens in doc]

print(tokens)
```

---

#### **4.6 Comparison Experiment**

```python
text = "I don't love NLP!!!"

print("NLTK:")
print(word_tokenize(text))

print("\nspaCy:")
doc = nlp(text)
print([tokens.text for tokens in doc])
```

1. **Why is tokenization considered the foundation of all NLP pipelines?**
Tokenization is considered the foundation of all NLP pipelines because it converts raw, continuous text into discrete, processable units called tokens, which are essential for any downstream computation. Machines cannot directly understand or operate on unstructured text like humans do; instead, they require structured inputs to perform tasks such as embedding generation, feature extraction, and model training. Every NLP system—whether classical models like TF-IDF or modern LLMs—operates on tokens rather than raw text. Without tokenization, it would be impossible to map text into numerical representations, making it a critical first step in enabling machines to understand and process language.

2. **What are token boundary problems? Give two examples.**
Token boundary problems refer to the challenge of determining where to correctly split text into tokens without losing semantic meaning or introducing ambiguity. These problems arise because natural language is complex and does not always follow clear separation rules. For example, in hyphenated words like “state-of-the-art” or “mother-in-law,” splitting them into separate tokens may break their intended meaning, while keeping them as one token may not always generalize well. Similarly, named entities such as “New York” or “Steve Jobs” represent single semantic units but are composed of multiple words, creating ambiguity in whether they should be treated as one token or multiple. These challenges directly impact model understanding and performance

3. **Why is spaCy is preferred over NLTK in production systems?**
spaCy is generally preferred over NLTK in production systems because it is designed for efficiency, scalability and real world deployment. It provides faster and more accurate tokenization by combining rule-based and statistical approaches, and it includes an optimized end to end NLP pipeline that handles sentence segmentation, tokenization, parts of speech tagging, and more in a unified framework. In contrast, NLTK is primarily an educational and research-oriented library that offers modular tools but requires more manual setup and is slower in execution. spaCy's production-ready design, better handling of edge cases, and ease of integration make it more suitable for building scalable NLP applications.
