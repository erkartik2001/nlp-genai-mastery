# 1. INTUITION
# =========================================

Think of raw text like messy, handwritten notes from different people:
- Some write in caps, some in lowercase
- Some use slang, emoji, punctuation
- Some repeat useless words like "this", "is", "and"

Before machine can understand this, we **clean and standardize it** just like organizing messy notes into a clean, readable format.

#### **What Problem does this solve ?**
- Raw text is inconsistent, noisy and ambiguous
- Models treat words literally - "Apple" != "apple"
- Noisy (punctuation, stopwords) can confuse models and increase computation

#### **Why was this needed?**
- Traditional NLP models (like TF-IDF, Naive Bayes) rely heavly on clean tokens
- Even modern systems benefit from structured input in certain pipelines.
- Helps:
    - Reduce vocabulary size
    - Improve model performance
    - speed up training and inference


# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 Text Processing (Definition)**

A sequence of steps applied to raw text to clean, normalize and convert into machine friendly format.

---
#### **2.2 Key Components**

#### **1. Text Normalization**
**Definition:** Converting text into standard format.

Includes:
- Lowercasing
- Removing extra spaces
- Expanding contractions (e.g., "don't" -> "do not")
---

#### **2. Lowercasing**
**Purpose:** Reduce duplicate tokens

Example:
- "Apple", "apple", "APPLE" -> all become "apple"
---

#### **3. Removing Punctuation**
**Purpose:** Removing symbols that don't add semantic meaning (in many tasks)

Example:
- "Hello!!!" -> "Hello"
**Exception:**
- Sentiment analysis (!!! matters)
- LLM prompts (punctuation affects meaning)
---

#### **4. Stopwords Removal**
**Definition:** Removing very common words with very low semantic value

Example:
- "the", "is", "in", "and"

Why:
- These words appear everywhere -> low information gain and high computing
---

#### **5. Stemming**
**Definition:** Reducing words to root form by chopping endings

Examples:
- "running" -> "run"
- "studies" -> "studi" (Not always correct)

Fast but linguistically accurate

---

#### **6. Lemmitization**
**Definition:** Converting words to their dictionary base form

Examples:
- "running" -> "run"
- "better" -> "good"

Its context aware, more accurate but slower than stemming.

---

#### **2.3 Step by Step working**
Pipeline example:

Raw Text: "I am Running!!! and I LOVE NLP."

Steps:
1. Lowercase -> "i am running!!! and i love nlp"
2. Remove Punctuation -> "i am running and i love nlp"
3. Tokenize -> ["i", "am", "running", "and", "i", "love", "nlp"]
4. Remove Stopwords -> ["running", "love", "nl;p"]
5. Lemmatize -> ["run", "love", "nlp"]

Final Output: ["run", "love", "nlp"]

# =========================================
# 3. INTERVIEW NOTES
# =========================================

#### **Key Points to Say**
- Text preprocessing reduces noise and improve model efficiency
- Important for traditional NLP pipelines
- Include normalization, tokenization, stemming, lemmitization etc

---

#### **Advantages**
- Reduces vocabulary size
- Improves model generalization
- Speed up computation
- Removes irrelavant information

---

#### **Disadvantages**
- Can remove important context
- Over-processing -> loss of meaning
- Not always needed for modern LLMs

---

#### **When to Use**
- Classical ML models (TF-IDF, BoW)
- search systems
- Text classification (basic pipelines)
- Keyword extraction

---

#### **When not to use**
- LLMs (like GPT, BERT) - they handle raw text well
- Tasks where punctuation matters (sentiment, sarcasm)
- Named entity recognition (case matters)
- Chatbots (tone matters)

---

#### **Important Comparisons**
**Stemming vs Lemmatization**

| **Feature** | **Stemming** | **Lemmatization** |
|-|-|-|
| Speed | Fast | Slower|
| Accuracy | Low | High|
| Output|Maybe Invalid| Valid word|
|Use case|Quick tasks|Producion systems|

# 4. IMPLEMENTATION
# =========================================

#### **4.1 Basic Setup**

```python
import re
import nltk
import spacy

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')
```

---
#### **4.2 Basic Preprocessing Pipeline**

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Lowercase
    text = text.lower()

    # Remove Punctuation
    text = re.sub(r'[^\w\s]','',text)

    # Tokenize 
    tokens = text.split()

    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization 
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return tokens
```

---

#### **4.3 Example**

```python
text = "I am Running!!! and i LOVE NLP."

print("Output:", preprocess_text(text))
```

---

#### **4.4 Stemming vs Lemmatization Experiment**

```python
words = ["running", "studies", "better"]

print("Stemming:")
print([stemmer.stem(w) for w in words])

print("Lemmatization:")
print([lemmatizer.lemmatize(w) for w in words])
```

---

#### **4.5 Using spaCy (Production-level)**

```python
nlp = spacy.load("en_core_web_sm")

def spacy_preprocess(text):
    doc = nlp(text)

    tokens = [
        (token.lemma_).lower() for token in doc
        if not token.is_stop and not token.is_punct
    ]

    return tokens

text = "I Love!!!! NLP"

print("Output:",spacy_preprocess(text))
```

# 6. MINI PROJECT / TASK
# =========================================

#### **Project: Build a text cleaning pipeline for spam detection**

**Input:**
A dataset of SMS messages: ["WIN a FREE lottery!!!", "HEY, are we metting today?"]

---

**Expected Output:**
Cleaned tokens: ["win", "free","lottery"], ["hey", "meeting", "today"]

```python
dataset = ["WIN a FREE lottery!!!", "HEY, are we meeting today?"]

def spam_detection(text):
    doc = nlp(text)

    tokens = [
        (token.text).lower() for token in doc 
        if not token.is_punct and not token.is_stop        
    ]

    return tokens

print("Output:")
for text in dataset:
    print(text," After preprocess : ",spam_detection(text))
```

#### 1. Why can removing stopwords hurt performance in some NLP tasks?
#### Ans: Stopword removal can hurt performance because stopwords often carry important contextual or structural meaning, especially in modern NLP tasks. 

#### Examples:
- Sentiment Analysis:
    - "not good" -> removing "not" will change meaning 
- RAG
    - Stopwords help preserving natural language flow imporving embedding quality
- Question-Answering:
    - Removal of stopwords breaks structure 
    
Key Insight: Stopwords are not always “useless”—they are low-frequency in meaning, but high in structure.

---

#### 2. When would you choose stemming over lemmatization?
#### Ans: In case of non production pipelines where fast preprocessing is critical, working with classical ML models (BoW, TF-IDF), exact word correctness is not important stemming will be used.

#### Examples:
    - Search engines (basic keyword matching)
    - Spam detection (fast pipelines)
    - Exploratory data analysis

Key Insight: Use stemming for speed, lemmatization for correctness or production systems.

---

#### 3. Why do modern LLMs not required heavy text processing?
#### Ans: Modern LLMs does not require heavy preprocessing because symbols which doesn't have any semantic are used as parts of input prompt for diving task user want in order or flow.

#### Reasons:
- Subword Tokenization
    - Techniques like BPE/WordPiece handle variations autmatical eg, running -> runs broken into meaningful subparts
- Contextual Understanding
    - Transformers understand meaning based on context not isolated words
- Trained on Raw text
    - LLMs are trained on massive, uncleaned corpora
    - They learn punctuation, casing and grammar naturally
- Preprocessing can remove signal
    - Removing punctuation, casing or stopwords can reduce meaning and intent

Key Insight: Traditional NLP manual cleaning required, Modern LLMs learn representation directly from raw data
