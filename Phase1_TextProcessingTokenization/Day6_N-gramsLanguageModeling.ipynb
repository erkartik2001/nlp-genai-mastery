# 1. INTUITION
# =========================================

Imagine you're typing on your phone and it suggests the next word:
- You type: **"I love"** -> it suggests "you", "coding", "food"

How does it know this?

It looks at **patterns in past text:**

- "I love you"" appears frequently
- "I love coding" appears frequently

This is exactly what **N-grams and language models** do.

#### **What problem does this solve?**

- Predict the **next word in a sequence**
- Understand **word relationships and context**
- Enable machines to generate text

#### **Why was this needed?**
- Early NLP systems coudn't understand context
- Needed a way to model:
    - Word order
    - Local dependencies
- N-grams provided a **simple statistical way** to model language

Key Idea: **Language is predictable based on previous words**

# =========================================
# 2. CORE CONCEPTS
# =========================================

#### **2.1 N-grams (Definition)**

A sequence of **N consecutive words (tokens)**

---

#### **2.2 Types of N-grams**

**Unigram (N=1)**

Single word

"I love NLP" -> ["I", "love", "NLP"]

---

**Bigram (N=2)**

Pairs of consecutive words

"I love NLP" -> ["I love", "love NLP"]

---

**Trigram (N=3)**

Triplets of consecutive words

"I love NLP" -> ["I love NLP"]

---

#### **2.3 N-gram Language Models**

**Definition**

A probablistic model that predicts the **next word based on previous N-1(minus) words** 

---

**Example (Bigram)

"I love NLP"

Model learns:

- P(love|I)
- P(NLP|love)

---

#### **2.4 Next Word Prediction**

**Idea**

Predict next word using probablitity 

Example: "I love __"

Possible Predictions

- "you" -> 0.5
- "coding" -> 0.3
- "food" -> 0.2

---

#### **2.5 Basic Probability Estimation**

**Bigram Probability**

P(word2 | word1) = Count(word1, word2) / Count(word1)

---

**Trigram Probability**

P(word3 | word1, word2) = Count(word1, word2, word3) / Count(word1, word2)

---

**Key Insight**

- Based on **frequency counts**
- Higher frequency -> higher probability

---

#### **2.6 Step-by-Step Working**

1. Tokenize corpus
2. Generate N-grams
3. Count frequencies
4. Calculate proabilities
5. Predict next word

Training:
Corpus -> N-grams -> counts -> probabilities

Inference:
Input -> lookup probabilities -> predict next word

Key Insight: N-gram models first learn probabilities from the corpus, and later use those stored probabilities to predict the next word.

# 4. IMPLEMENTATION
# =========================================

#### **4.1 Setup**

```python
import nltk
from nltk.util import ngrams
from collections import Counter

nltk.download("punkt")
```

---

#### **4.2 Generate N-grams**

```python
text = "I love NLP and I love AI"
tokens = nltk.word_tokenize(text)
# print(tokens)

bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("Bigrams: ",bigrams)
print("Trigrams: ",trigrams)
```

---

#### **4.3 Frequency Count**

```python
bigrams_counts = Counter(bigrams)
print(bigrams_counts)
```

---

#### **4.4 Probability Estimation**

```python
def calculate_bigram_prob(bigrams_counts, tokens):
    unigram_counts = Counter(tokens)
    # print(unigram_counts)

    probs = {}

    for (w1,w2), count in bigrams_counts.items():
        probs[(w1,w2)] = count / unigram_counts[w1]

    return probs

probs = calculate_bigram_prob(bigrams_counts, tokens)
print(probs)
```

---

#### **4.5 Next Word Prediction**

```python
def predict_next(word, bigrams_probs):
    candidates = {
        w2: prob for (w1, w2), prob in bigrams_probs.items() if w1==word
    }

    return sorted(candidates.items(), key = lambda x: x[1], reverse=True)


print(predict_next("love",probs))
```

---

#### **4.6 Variation: Trigram Model**

```python
trigram_counts = Counter(trigrams)

trigram_counts = Counter(trigrams)

def predict_next_trigram(w1, w2, trigram_counts):
    candidates = {
        w3: count for (x1, x2, w3), count in trigram_counts.items()
        if x1 == w1 and x2 == w2
    }
    
    return sorted(candidates.items(), key=lambda x: x[1], reverse=True)
```

# 8. SELF-TEST QUESTIONS
# =========================================

1. **Why do N-gram models suffer from data sparsity?**

N-gram models suffers from data sparsity because as the value of N increases, the number of possible word combinations grows exponentially, while the available training data remains limited. This means many valid word sequences either appear very rarely or do not appear at all in the corpus, making it difficulrt to estimate reliable probabilities for them. Since N-gram models rely purely on frequency counts of observed sequecnes and not understand semantic meaning, they cannot generalize to unseen or rare combinations. As a result, the model assings zero or very low to many valid sequecnes, leading to poor performance. This problem becomes more severe with order N grams like trigrams or beyond.


2. **How does increasing N improve and also hurt performance?**

Increasing N improves performance by incorporating more context into predictions, allowing the model to better capture local word dependencies and generate more accurate next word predictions. For example a trigram model considers the previous 2 words, which provides more contextual information than a bigram model. However increasing N significantly increases the number of possible word combinations, leading to data sparsity and unreliable probability estimates, espcially when the dataset is not large enough. Additionally higher N increases memory usage and computational complexity. Therefore while larger N can improve accuracy in theory, in practice it often hurts the performance due to sparse data and scalability issues.


3. **How are LLMs fundamentally different from N-gram in next word prediction?**

N-gram models and LLMs both aim to predict the next word, but they differ fundamentally in how they model language. N-gram models are based on simple frequency counts and estimate probabilities using fixed-length context windows, without understanding semantics or long-range dependencies. In contrast, LLMs use deep learning architectures, specifically transformers, to learn contextual representations of language from large-scale data. They can capture long-range dependencies, understand semantic relationships, and generalize to unseen patterns. Instead of relying on exact matches of previous word sequences, LLMs learn distributed representations (embeddings) that allow them to predict the next word based on deeper contextual understanding.
