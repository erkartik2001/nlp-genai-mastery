#### **2.1 Unicode Normalization**
**Definition**

#### **Unicode** 
Unicode is a universal encoding system that assigns a unique code to every character across languages and symbols, so computers can represent text consistently. For example, A is U+0041, अ is U+0905, and 😊 is U+1F60A. Instead of storing letters directly, systems store these codes, which ensures that text looks the same across devices and platforms.

#### **Unicode Normalization**
Unicode normalization ensures that visually identical characters are stored in a consistent way, because some characters can have multiple internal representations. For example, é can be a single character (U+00E9) or a combination of e + ́. Without normalization, "é" == "é" returns False. Using normalization like NFC (Normalization Form Composed) combines characters into a single standard form, while NFD (Normalization Form Decomposed) splits them into base character + accent. This helps in accurate text matching, searching, and NLP processing.


Standardizing text representation so that **visually identical characters are treated the same**

Example:
- é" can be:
    - Single character (NFC)
    - 'e' + accent (NFD)

#### **Why needed?**
- Prevent duplicate representations
- Ensure consistent tokenization

#### **Types**
- NFC (composed) -> preferred
- NFD (decomposed)

**Problem**
- Emjois carry **strong semantic meaning**
- Example:

**Strategies**
1. Remove emojis (lose meaning)
2. Replace with text ("😊" → "smile")

#### **2.6 Regex for Text Processing**
**Definition**
A powerfull pattern matching tool for **searching(find patterns), replace(clean text), extract(pull useful info) or overall transform text**

**Common Patterns**

1. **URLs (r'https?://\S+') What it checks:**
    - http or https (s? = optional s)
    - :// literal
    - \S+ -> everything until space
    - Matches eg. https://google.com, http://zxy.in/page?id=1
    - Use to replace it with < URL > or extract links

2. **Numbers (r'\d+') What it checks:**
    - \d -> digits (0-9)
    - + -> one or more
    - Matches eg. 123, 2026, 45
    - Use to remove numbers or replace with < NUM >

3. **Punctuation (r'[^\w\s]') What it checks:**
    - \w -> letters, digits, _
    - \s -> spaces
    - ^ -> NOT
    - Match anything that is not word or space
    - Matches: ! @ # . , ? ' "
    - Use to remove punctuation

4. **Extra Spaces (r'\s+') What it checks:**
    - \s -> white spaces
    - +-> multiple
    - Matches eg. "Hello   world"
    - Use to normalize spacing

5. **Emails (r'\S+@\S+\.\S+') What it checks:**
    - text + @ + domain + .com/.in
    - Matches eg. text@gmail.com, user@yahoo.com
    - Use to replace with < EMAIL >

6. **Phone Numbers (r'\d{10}') What it checks:**
    - Exactly 10 digits
    - Matches 9876543210

7. **Hashtags (r'#\w+') What it checks:**
    - #followed by word
    - Matches eg. #AI #MachineLearning

8. **Mentions (r'@\w+') What it checks:**
    - @ followed by username
    - Matches @john @openai

9. **Repeated Punctuation (r'[!?.]+') What it checks:**
    - Multiple !?.
    - Matches !!! ??? ...
    - Use to replace -> single !

10. **Words only (r'\b\w+\b') What it checks:**
    - \ b -> word boundary
    - \w+ -> full word
    - Matche eg. hello, world -> ["hello", "world"]

11. **Non-ASCII(removing emojis/special chars) (r'[^\x00-\x7F]+') What it checks:**
    - Characters outside standard ASCII
    - Matches eg. 😊🔥💯


**Pattern Actions**

1. **Search (check if exists):** 
    - re.search(pattern, text)
    - Returns match if pattern exists

2. **Extract:**
    - re.findall(pattern, text)
    - Return all matches as list

3. **Replace:**
    - re.sub(pattern, replacement, text)
    - Replace matched patterns

# 4. IMPLEMENTATION
# =========================================

#### **4.1 Setup**

```python
import re
import unicodedata
```

---

#### **4.2 Unicode normalization**

```python
def normalize_unicode(text):
    return unicodedata.normalize('NFC', text)
```

---

#### **4.3 Regex Cleaning Functions**

```python
def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '<URL>', text)

def replace_numers(text):
    return re.sub(r'\d+','<NUM>', text)

def remove_special_chars(text):
    return re.sub(r'[^\w\s<>]','',text)

def clean_spaces(text):
    return re.sub(r'\s+',' ',text).strip()
```

---

#### **4.4 Emoji Handling (Basic)**

```python
def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        "]",
        flags=re.UNICODE
    )
    return emoji_pattern.sub('', text)
```

---

#### **4.5 Full Pipeline**

```python
def preprocess_advanced(text):
    text = normalize_unicode(text)
    text = text.lower()
    text = remove_urls(text)
    text = replace_numers(text)
    text = remove_emojis(text)
    text = remove_special_chars(text)
    text = clean_spaces(text)
    
    return text
```

---

#### **4.6 Example**

```python
text = "I love NLP 😊! Visit https://example.com, price is ₹1000!!!"

print(preprocess_advanced(text))
```

---

#### **4.7 Variation: Keep Emojis as Meaninig**

```python
def replace_emojis_simple(text):
    return text.replace("😊", "smile").replace("😢", "sad")

print(replace_emojis_simple(text))
```

# 6. MINI PROJECT / TASK
# =========================================

#### **Project: Clean Social Media Text**
**Input**

[
 "OMG 😍 this is amazing!!! Visit http://abc.com",
 "Price is ₹5000 😡 #expensive"
]

---

#### **Expected Output**

[
 "omg this is amazing visit <URL>",
 "price is <NUM> expensive"
]

```python
dataset = [
 "OMG 😍 this is amazing!!! Visit http://abc.com",
 "Price is ₹5000 😡 #expensive"
]

def remove_urls(text):
    return re.sub(r'https?://\S+|www\.\S+', '<URL>', text)

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"
        u"\U0001F680-\U0001F6FF"
        "]",
        flags=re.UNICODE
    )
    return emoji_pattern.sub('', text)
    
def remove_special_characters(text):
    return re.sub(r'[^\w\s<>]','',text)

def remove_numbers(text):
    return re.sub('\d+','<NUM>',text)

def preprocess_text(text):
    text = text.lower()
    text = remove_urls(text)
    text = remove_numbers(text)
    text = remove_emojis(text)
    text = remove_special_characters(text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text

for text in dataset:
    print(preprocess_text(text))

```

#### **1. Why is replacing URLs with <URL> often better than removing them completely?**
Replacing actual urls with <URL> tag helps in maintaing the context of the textual dataset, the model will peform better with that type of data instead of getting confused what actually has been followed after particular sentence. Ex. Instead of "Get instant discount using this  ", "get instand discount using this < URL >". In the second example, the text has more context that on a particular url a discount is been given.


#### **2. How can emojis improve performance in sentiment analysis tasks?**
There are multiple types of emjois expressing different emotions eg. happy, sad, cry, laugh. Using that data user intent or emotion expression in prompt input or text dataset can be understood. For example in a sentence "i like you", without the user is expressing his her relationship with any entity or a person to a particular extent. But if emoji like heart or lips added after that, it adds more xontext that the user has more than liking to loving or caring or genuine love.


#### **3. What are the risks of overusing regex in preprocessing pipelines?**
Over using of regex in preprocessing piplines, first simply can lose a lot of context, also different lanugages can have differnt regex pattern which eventually leads to more hard regex pattern.
